### Atividade 06:
#### Crie um gráfico para avaliar a tendência dos sequestros relâmpago.
<br>

---

```python
# Visualizando a tendência temporal de homicídios dolosos
plt.figure(figsize=(15,5))
sns.lineplot(x=df['ano'], y=df['sequestro_relampago'], marker='o')
plt.title("Evolução dos sequestros relâmpago no Rio de Janeiro")
plt.xlabel("Ano")
plt.ylabel("Sequestros relâmpago")
plt.grid()
plt.show()
```

```python
# Visualizando a tendência temporal de pessoas desaparecidas por mês
plt.figure(figsize=(15,5))
sns.lineplot(x=df['mes'], y=df['pessoas_desaparecidas'], marker='o')
plt.title("Evolução dos sequestros relâmpago no Rio de Janeiro")
plt.xlabel("Mês")
plt.ylabel("Pessoas desaparecidas")
plt.grid()
plt.show()
```

### Atividade 11:
#### Crie um gráfico para exibir os homicídios dolosos ao longo dos anos.
<br>

---

```python
# Comparando crimes letais ao longo dos anos
plt.figure(figsize=(12,5))
sns.barplot(x=df['ano'], y=df['hom_doloso'], hue=df['ano'], palette='Reds', legend=False)
plt.title("Homicídios Dolosos por ano entre 2003 e 2024")
plt.xlabel("Ano")
plt.ylabel("Número de Homicídios")
plt.xticks(rotation=45)
plt.show()
```

```python
# Comparando crimes letais ao longo dos anos
plt.figure(figsize=(12,5))
sns.barplot(x=df['ano'], y=df['hom_doloso'], hue=df['ano'], width=0.5)
plt.title("Homicídios Dolosos por ano entre 2003 e 2024")
plt.xlabel("Ano")
plt.ylabel("Número de Homicídios")
plt.xticks(rotation=45)
plt.show()
```

## Conclusão

A análise dos dados revelou que os índices de segurança no Rio de Janeiro oscilaram ao longo dos anos, trazendo algumas percepções importantes:
	•	Sequestros relâmpago: Os casos tiveram altos e baixos ao longo do tempo, o que pode estar ligado a mudanças na segurança pública, estratégias de combate ao crime ou até mesmo a novas formas de atuação dos criminosos.
	•	Desaparecimentos de pessoas: Os dados apontam para padrões que se repetem em determinados meses e anos, sugerindo que fatores como condições sociais, econômicas ou até mesmo épocas do ano podem influenciar esses números.
	•	Homicídios dolosos: As taxas subiram e desceram em diferentes períodos, o que pode refletir o impacto de políticas de segurança, operações policiais e até mudanças no próprio comportamento da sociedade.

Essas variações mostram que a segurança pública é um tema complexo, influenciado por diversos fatores. Compreender essas tendências pode ajudar a direcionar estratégias mais eficazes para proteger a população.
